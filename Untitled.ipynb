```python
#!/usr/bin/env python3

import pandas as pd
import matplotlib.pyplot as plt
import argparse
# import sys
# import signal

# RED = "\033[91m"
# GREEN = "\033[92m"
# YELLOW = "\033[93m"
# RESET = "\033[0m"

## NEED PARSING
# === PARSING === #
def parse_arguments():
	parser = argparse.ArgumentParser()

	parser.add_argument("--file", type=str, help='Path to the CSV file')

	return parser.parse_args()

# === GET DATAS === #
def get_data_values(args):
	try:
		if args.file:
			csv_path = args.file
			data = pd.read_csv(csv_path)
			return data.iloc[:, 0], data.iloc[:, 1]
		else:
			raise ValueError(f"{RED}Please provide a path to the CSV file using the --file argument.{RESET}")
	except e as Exception:
		print(f"{e}")

def get_divider(nb):
	divider = 1
	while not 0 < nb / divider < 1:
		divider *= 10
	return divider

def normalize(list):
	larger_val = max(abs(list))
	if not 0 <= larger_val <= 1:
		divider = get_divider(larger_val)
		list /= divider

	return list, divider

# === VISUALIZE DATAS === #
def scatter_dataset(X, Y):
	try:
		plt.scatter(X, Y, label='value', color="blue", marker='o')

		plt.xlabel('Kms')
		plt.ylabel('Price')
		plt.title('data.csv representation')

		plt.show()
	except e as Exception:
		print(F"{e}")

def predictions_line_representation(X, Y, Y_pred):
	try:
		# Show the plot
		plt.plot(X, Y_pred, color= "red")
		scatter_dataset(X, Y)
	except e as Exception:
		print(f"{e}")

def training_vizualisation(epochs, costs):
	try:
		plt.xlabel('Iterations')
		plt.ylabel('Cost')
		plt.title('Cost Function over Iterations')
		plt.plot(epochs, costs, label='Cost Function')
		plt.show()
	except e as Exception:
		print(f"{e}")

# === MATHS FUNCTIONS === #
def get_Ypredictions(a, X, b):
	return a * X + b

def cost_function(X, Y, n, a, b):
	## Mean Square Error
	cost = (1 / (2 * n)) * (sum((Y - (a * X + b)) ** 2))
	return cost

def gradient_descent(X, Y, n, a, b, L):
	Y_pred = get_Ypredictions(a, X, b)
	# calculer la derivee de notre courbe
	Da = (1 / n) * sum(X * (Y_pred - Y))
	Db = (1 / n) * sum(Y_pred - Y)
	# calculer les valeurs a(i + 1) et b(i + 1)
	# == les valeurs de a et b au step suivant
	a_next = a - L * Da
	b_next = b - L * Db
	return a_next, b_next

# === PREDICTION === #
# def try_preditcion(a, b, x_divider, y_divider):
# 	c = ""
# 	val = ""
# 	while not c in ["n"]:
# 		c = input(f"{YELLOW}Do you want to try a prediction before training ? [Y/n/exit]:{RESET}")
# 		if c in ["Y", "y"]:
# 			while not val.isdigit():
# 				val = input(f"How many kilometers does your vehicle have ?:")
# 			prediction = (a * (int(val) / x_divider) + b) * y_divider # denormalized value
# 			val = ""
# 			if prediction < 0:
# 				print(f"{RED}Model has predicted a negative value.\nYou should not sell your vehicle...{RESET}")
# 			else:
# 				print(f"Estimated price:{abs(prediction)}")

# === MAIN === #
# def signal_handler(sig, frame):
# 	print(f"{RED}Signal detected.{RESET}")
# 	sys.exit(1)

if __name__ == "__main__":
	# signal.signal(signal.SIGINT, signal_handler)

	args = parse_arguments()

	X, Y = get_data_values(args)
	scatter_dataset(X, Y)
	X, x_divider = normalize(X)
	Y, y_divider = normalize(Y)
	# initialize parameters
	a		= 0				# slope
	b		= 0				# intercept
	n		= len(X)		# n values in dataset
	L		= 0.1			# learning rate
	epochs	= 10000			# number of iterations
	costs	= [0] * epochs	# cost evolution during iterations
	
	# try prediction before training
	# try_preditcion(a, b, x_divider, y_divider)

	## iterative - gradient descent
	for i in range(epochs):
		costs[i] = cost_function(X, Y, n, a, b)
		a, b = gradient_descent(X, Y, n, a, b, L)
	print(f"{GREEN}Your model is trained !{RESET}")

	## visualize
	predictions_line_representation(X * x_divider, Y * y_divider, get_Ypredictions(a, X, b) * y_divider)
	# training_vizualisation(epochs, costs)

	# try prediction after training
	# try_preditcion(a, b, x_divider, y_divider)
```
